```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np
```

```python
df = pd.read_csv('../data/hotel_booking_clean.csv')
df.head(5)
```

# Multivariate Analysis (with ~cancellations/cancellations)

## Lead Time vs ADR

```python
df[df['adr'] < 0]['adr']
```

```python
df = df[(df['adr']>=0) & (df['adr'] < 5000)]
```

```python
labels = [
    'Very Low',
    'Low',
    'Medium',
    'High',
    'Very High'
]

df['lead_time_bin'] = pd.qcut(df['lead_time'], q=5, labels = labels)
df['adr_bin'] = pd.qcut(df['adr'], q=5, labels = labels)

heatmap_data = pd.pivot_table(
    df,
    values='is_canceled',
    index='lead_time_bin',
    columns='adr_bin',
    aggfunc='mean'
)
```

```python
fig, ax = plt.subplots(figsize=(10,6))

im = ax.imshow(heatmap_data, cmap = 'Blues')

colour_bar = plt.colorbar(im)

colour_bar.set_label('Probability of Cancellation Rate')

ax.set_xticks(range(len(heatmap_data.columns)))
ax.set_xticklabels(heatmap_data.columns, rotation=45)
ax.set_xlabel('ADR')
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

ax.set_yticks(range(len(heatmap_data.index)))
ax.set_yticklabels(heatmap_data.index)
ax.set_ylabel('Lead Time')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

ax.scatter(
    df[df['is_canceled']==0]['lead_time'],
    df[df['is_canceled']==0]['adr'],
    alpha=0.3,
    label='Not Cancelled'
)

ax.scatter(
    df[df['is_canceled']==1]['lead_time'],
    df[df['is_canceled']==1]['adr'],
    alpha=0.3,
    label='Cancelled'
)

ax.set_xlabel('Lead Time')
ax.set_ylabel('ADR')

ax.legend()

plt.show()
```

## Distribution Channel vs Market Segment vs Customer Type

```python
fig, ax = plt.subplots(figsize=(10,5))

pd.crosstab(df['market_segment'],df['distribution_channel'],normalize='index').plot(kind='bar', ax = ax)

ax.set_ylabel('Proportion of Bookings')
ax.legend(bbox_to_anchor= (1.05,1),loc='upper left')

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(10,5))

ct = pd.crosstab(
    [df['market_segment'], df['distribution_channel']],
    df['customer_type'],
    normalize='index'
)

ct.plot(
    kind='bar',
    stacked=True,
    figsize=(12,6),
    ax=ax
)

ax.legend(bbox_to_anchor= (1.05,1),loc='upper left')

plt.ylabel('Proportion of Customer Type')
plt.show()
```
