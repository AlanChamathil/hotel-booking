# Reusable Functions

```python
def reusable_bar_chart_plot(feature_name):
    fig, ax = plt.subplots(2,1, figsize=(20,10))
    
    x_values = df[feature_name].value_counts().sort_index().values
    y_values = df[feature_name].value_counts().sort_index().index
    
    bars = ax[0].bar(y_values,x_values, color='skyblue')
    
    for bar in bars:
        height = bar.get_height()
        ax[0].text(
            bar.get_x() + bar.get_width()/2,
            height/2,
            f"{int(height)}",
            ha = "center",
            va = "bottom"
        )
    
    ax[0].set_title(feature_name)
    ax[0].set_xlabel(feature_name)
    ax[0].set_ylabel('Bookings')
    ax[0].set_xticks(df[feature_name].value_counts().sort_index().index)
    x_values_percent = df[feature_name].value_counts(normalize=True).sort_index().values*100
    
    bars = ax[1].bar(y_values,x_values_percent, color='skyblue')
    
    for bar in bars:
        height = bar.get_height()
        ax[1].text(
            bar.get_x() + bar.get_width()/2,
            height/2,
            f"{height:.1f}%",
            ha = "center",
            va = "bottom"
        )
    
    ax[1].set_title(feature_name+' in %')
    ax[1].set_xlabel(feature_name)
    ax[1].set_ylabel('Bookings %')
    ax[1].set_xticks(df[feature_name].value_counts().sort_index().index)
    
    
    plt.show()
```

### Inference on lead time (Histogram)

Seems like most of the observation is concertrated between 0 and 100. However, 0 - 14 lead time seems to be most prevalent. This infers most individuals booked on the same day of the booking,. 

* Heavily right-skewed (positively skewed)
* most values are clustered on the left (low lead time)
* long tail stretching to the right (high lead time)

Most customers book close to their arrival date, but a smaller group books far in advance
A large proportion of bookings are made within ~0-50 days before arrival.
Some customers book extremely early (1-2 years), but these are rare

The distribution of lead time is heavily right-skewed, with the majority of bookings occurring within a short period prior to arrival. The frequency of bookings decreases as lead time increases, indicating that early bookings are less common. A long tail is observed, with some bookings made several hundred days in advance, suggesting the presence of extreme values. This feature is likely to be important in predicting cancellations, as longer lead times may be associated with higher uncertainty and thus a higher likelihood of cancellation.

```python
fig, ax = plt.subplots(figsize = (20,5))

ax.boxplot(df['lead_time'],  vert=False)
ax.set_title("Lead Time Boxplot")
ax.set_xlabel("Lead time (Days)")
plt.show()
```

## Arrival Year

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

x_values = df['arrival_date_year'].value_counts().sort_index().values
y_values = df['arrival_date_year'].value_counts().sort_index().index

bars = ax[0].bar(y_values, x_values, color=['skyblue','steelblue','dodgerblue'])

for bar in bars:
    height = bar.get_height()
    ax[0].text(
        bar.get_x() + bar.get_width()/2,
        height/2,
        f"{int(height)}",
        ha = "center",
        va = "bottom"
    )

ax[0].set_title("Arrival Year")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Bookings")

ax[0].set_xticks(y_values)

x_values_percent = df['arrival_date_year'].value_counts(normalize=True).sort_index().values*100
bars = ax[1].bar(y_values, x_values_percent, color=['skyblue','steelblue','dodgerblue'])

for bar in bars:
    height = bar.get_height()
    ax[1].text(
        bar.get_x() + bar.get_width()/2,
        height/2,
        f"{height:.1f}%",
        ha = "center",
        va = "bottom"
    )

ax[1].set_title("Arrival Year in %")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Bookings %")

ax[1].set_xticks(y_values)

plt.show()
```

* 2016 47.5% bookings were made
* 2017 secondd most bookings made 34.2%
* 2015 the least bookings made 18.4%

## Arrival Month

```python
fig, ax = plt.subplots(2,1, figsize=(15,10))

month_order = [
    'January', 'February', 'March', 'April',
    'May', 'June', 'July', 'August',
    'September', 'October', 'November', 'December'
]

x_values = df['arrival_date_month'].value_counts().reindex(month_order).values
y_values = month_order

bars = ax[0].bar(y_values,x_values, color='skyblue')

for bar in bars:
    height = bar.get_height()
    ax[0].text(
        bar.get_x() + bar.get_width()/2,
        height/2,
        f"{int(height)}",
        ha = "center",
        va = "bottom"
    )

ax[0].set_title('Arrival Month')
ax[0].set_xlabel('Months')
ax[0].set_ylabel('Bookings')

x_values_percent = df['arrival_date_month'].value_counts(normalize=True).reindex(month_order).values*100

bars = ax[1].bar(y_values,x_values_percent, color='skyblue')

for bar in bars:
    height = bar.get_height()
    ax[1].text(
        bar.get_x() + bar.get_width()/2,
        height/2,
        f"{height:.1f}%",
        ha = "center",
        va = "bottom"
    )

ax[1].set_title('Arrival Month in %')
ax[1].set_xlabel('Months')
ax[1].set_ylabel('Bookings %')



plt.show()
```

* seems like bookings are popular thoughout Feb to Oct
* July and August stands out the most popular months to book. Which make sense, since it summer period
* Jan, Nov, and Dec the least booked which make sense, since it's winter period
