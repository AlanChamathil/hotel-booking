```python
import pandas as pd
```

```python
data = [[1, 3.5], [2, 3.65], [3, 4.0], [4, 3.85], [5, 4.0], [6, 3.65]]
scores = pd.DataFrame(data, columns=['id', 'score']).astype({'id':'Int64', 'score':'Float64'})
```

```python
df = scores.sort_values(by='score', ascending=False)
```

```python
df
```

```python
df['score']
```

```python
df['rank'] = df.sort_values(by='score', ascending=False).rank(method='dense', ascending=False)['score'].astype(int)
```

```python
df.drop(columns=['id']).reset_index(drop=True)
```

```python
pd.DataFrame({
    df['score']: [df['score'].rank(method = 'dense', ascending=False).astype(int)]
})
```
